### Missing Value

```python
df.isnull().sum()
```

Fill missing value of Glucose, BloodPressure, and BMI with mean (have normal distribution)

```python
for i in ['Glucose','BloodPressure','BMI']:
  df[i].fillna(df[i].mean(), inplace=True)
```

Fill Missing Values of Insulin and SkinThickness by using KNN method

```python
df_KNN = pd.get_dummies(df[["Insulin","SkinThickness"]], drop_first=True)
```

```python
# Standardization Variables
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_KNN = pd.DataFrame(scaler.fit_transform(df_KNN), columns=df_KNN.columns)
df_KNN.head()
```

```python
# Implement KNN Method
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5)

df_KNN = pd.DataFrame(imputer.fit_transform(df_KNN), columns=df_KNN.columns)
df_KNN.head()
```

```python
# Undo Standardization
df_KNN = pd.DataFrame(scaler.inverse_transform(df_KNN), columns=df_KNN.columns)
df_KNN.head()
```

```python
df["Insulin"] = df_KNN["Insulin"]
df["SkinThickness"]= df_KNN["SkinThickness"]
```

### Outliers

```python
# Inter Quartile Range (IQR)
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
low_limit = Q1 - (IQR * 1.5)
high_limit = Q3 + (IQR * 1.5)

outliers = ((df < low_limit ) | (df > high_limit))
outliers.sum()
```

```python
for column in df.columns:
        plt.figure(figsize=(10,1))
        sns.boxplot(data=df[df.columns], x=column)
```

```python
# Outliers > Median
df_OutMed = df[['Pregnancies']]
for i in df_OutMed.columns:
  median = df_OutMed.median()
  df_OutMed.mask(outliers, other=median, axis=1, inplace=True)
```

```python
# Outliers > Mean
df_OutMean = df[['SkinThickness', 'Insulin','DiabetesPedigreeFunction']]
for i in df_OutMean.columns:
  mean = df_OutMean.mean()
  df_OutMean.mask(outliers, other=mean, axis=1, inplace=True)
```

```python
df['Pregnancies'] = df_OutMed['Pregnancies']
df['SkinThickness'] = df_OutMean['SkinThickness']
df['Insulin'] = df_OutMean['Insulin']
df['DiabetesPedigreeFunction'] = df_OutMean['DiabetesPedigreeFunction']
```

```python
for column in df.columns:
        plt.figure(figsize=(10,1))
        sns.boxplot(data=df[df.columns], x=column)
```

## Feature Engineering

### Classifying

```python
Age_Class = []
for i in df["Age"]:
    if 20 < i <= 30:
        Age_Class.append("Young")  
    elif 30 < i <= 44:
        Age_Class.append("Middle")  
    elif 44 < i <= 60:
        Age_Class.append("Adult")                     
    elif i > 50 :
        Age_Class.append("Elderly")    
    else:
        Age_Class.append(np.NaN)
```

```python
df['Age_Class'] = Age_Class
```

```python
def check (col):
  a = df[col].value_counts()
  b = df[col].value_counts(normalize=True)*100
  result = pd.concat([a,b], axis = 1, keys=['Counts','%'])
  return result
```

```python
check('Age_Class')
```

```python
Glucose_Class = []
for i in df['Glucose']:
  if i < 70:
    Glucose_Class.append('Hipoglisemi')
  elif 70 <= i < 100:
    Glucose_Class.append('Normal')
  elif 100 <= i < 126:
    Glucose_Class.append('Imparied')
  elif i >= 126 :
    Glucose_Class.append('Hiperglisemi')
  else:
        Glucose_Class.append(np.NaN)
```

```python
df['Glucose_Class'] = Glucose_Class
```

```python
check('Glucose_Class')
```

```python
BMI_Class = []
for i in df["BMI"]:
    if i < 16:
        BMI_Class.append("Over Weak")  
    elif 16 <= i < 18.5:
        BMI_Class.append("Weak")  
    elif 18.5 <= i < 25:
        BMI_Class.append("Normal")
    elif 25 <= i < 30:
        BMI_Class.append("Over Weight")
    elif 30 <= i < 35:
        BMI_Class.append("1st Obese")
    elif 35 <= i < 45:
        BMI_Class.append("2nd Obese")                      
    elif i >= 45 :
        BMI_Class.append("3rd Obese")    
    else:
        BMI_Class.append(np.NaN)
```

```python
df['BMI_Class'] = BMI_Class
```

```python
check('BMI_Class')
```

```python
BloodPressure_Class = []
for i in df['BloodPressure']:
  if i < 70:
    BloodPressure_Class.append('Low')
  elif 70 <= i < 90:
    BloodPressure_Class.append('Normal')
  elif i >= 90 :
    BloodPressure_Class.append('High')
  else:
        BloodPressure_Class.append(np.NaN)
```

```python
df['BloodPressure_Class'] = BloodPressure_Class
```

```python
check('BloodPressure_Class')
```

```python
Insulin_Class = []
for i in df['Insulin']:
  if i < 120:
    Insulin_Class.append('Normal')
  elif i >= 120 :
    Insulin_Class.append('Abnormal')
  else:
        Insulin_Class.append(np.NaN)
```

```python
df['Insulin_Class'] = Insulin_Class
```

```python
Pregnancies_Class = []
for i in df['Pregnancies']:
  if i < 1:
    Pregnancies_Class.append('Un-Pregnant')
  elif 1 <= i <= 5:
    Pregnancies_Class.append('Normal')
  elif 5 < i <= 10:
    Pregnancies_Class.append('High')
  elif i >= 10 :
    Pregnancies_Class.append('Very High')
  else:
        Pregnancies_Class.append(np.NaN)
```

# Machine Learning

## Train-Test Split

```python
X = new_df
X.shape
```

```python
y = target
y.shape
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X, y, test_size = 0.3, random_state = 123, stratify=y)
```

## Resampling SMOTE

```python
y_train.value_counts()
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
# define undersampling strategy
oversample = SMOTE(sampling_strategy=1)

# fit and apply the transform
X_train_smote, y_train_smote = oversample.fit_resample(X_train, y_train)
```

```python
y_train_smote.value_counts()
```

## Model

```python
# membuat fungsi evaluasi model
def evaluasi_model(model, X_test, y_test):
    y_pred = model.predict(X_test)
    return accuracy_score(y_test, y_pred)
```

### Decision Tree

```python
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier(max_depth=5, random_state=12)
dtc.fit(X_train_smote, y_train_smote)
```

```python
train_dt = evaluasi_model(dtc, X_train_smote, y_train_smote)
train_dt
```

```python
test_dt = evaluasi_model(dtc, X_test, y_test)
test_dt
```

```python
# Mencari depth terbaik
from sklearn.model_selection import GridSearchCV

max_depth = np.array([1,2,3,4,5])
param_grid = dict(max_depth=max_depth)
grid = GridSearchCV(estimator=dtc, param_grid=param_grid)
grid.fit(X_train_smote, y_train_smote)
print(grid.best_score_)
print(grid.best_estimator_.max_depth)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
```

```python
# Find Confusion Metrix
y_pred_dt = dtc.predict(X_test)
print(classification_report(y_test, y_pred_dt))
```

```python
from sklearn.metrics import confusion_matrix

# Order of the input parameters is important: 
# first param is the actual output values
# second param is what our model predicted
dt_conf_matrix = confusion_matrix(y_test, y_pred_dt)

dt_conf_matrix
```

```python
%%writefile streamlit_app_pred.py
import pandas as pd
import pickle
import streamlit as st
from xgboost import XGBClassifier

with open("lxgb_model.pkl","rb") as file:
    classifier = pickle.load(file)

@st.cache()
# Function for user input
def prediction_(Pregnancies_Class, Glucose_Class, BloodPressure_Class, SkinThickness, Insulin_Class, BMI_Class, DiabetesPedigreeFunction, Age_Class):

  SkinThickness = SkinThickness
  DiabetesPedigreeFunction = DiabetesPedigreeFunction

  if Age_Class == 'Young':
    Age_Class = 1 
  elif Age_Class == 'Middle':
    Age_Class = 2 
  elif Age_Class == 'Adult':
   Age_Class = 3                     
  elif Age_Class == 'Elderly' :
    Age_Class = 4    
  else:
    Age_Class = 5

  if Glucose_Class == 'Hipoglisemi':
    Glucose_Class = 1
  elif Glucose_Class == 'Normal':
    Glucose_Class = 2
  elif Glucose_Class == 'Imparied':
    Glucose_Class = 3
  elif Glucose_Class == 'Hiperglisemi' :
    Glucose_Class = 4
  else:
    Glucose_Class = 5

  if BMI_Class == 'Over Weak':
    BMI_Class = 1  
  elif BMI_Class == 'Weak':
    BMI_Class = 2  
  elif BMI_Class == 'Normal':
    BMI_Class = 3
  elif BMI_Class == 'Over Weight':
    BMI_Class = 4
  elif BMI_Class == '1st Obese':
    BMI_Class = 5
  elif BMI_Class == '2nd Obese':
    BMI_Class = 6                      
  elif BMI_Class == '3rd Obese' :
    BMI_Class = 7   
  else:
    BMI_Class = 8

  if BloodPressure_Class =='Low':
    BloodPressure_Class = 1
  elif BloodPressure_Class == 'Normal':
    BloodPressure_Class = 2
  elif BloodPressure_Class == 'High':
    BloodPressure_Class = 3
  else:
    BloodPressure_Class = 4

  if Insulin_Class == 'Normal':
    Insulin_Class = 1
  elif Insulin_Class == 'Abnormal':
    Insulin_Class = 2
  else:
    Insulin_Class = 3

  if Pregnancies_Class == 'Un-Pregnant':
    Pregnancies_Class = 1
  elif Pregnancies_Class == 'Normal':
    Pregnancies_Class = 2
  elif Pregnancies_Class == 'High':
    Pregnancies_Class = 3
  elif Pregnancies_Class == 'Very High':
    Pregnancies_Class = 4
  else:
    Pregnancies_Class = 5

  pred = classifier.predict([[Pregnancies_Class, Glucose_Class, BloodPressure_Class, SkinThickness, Insulin_Class, BMI_Class, DiabetesPedigreeFunction, Age_Class]])

  return pred

  def main():
  html_temp = """
  <div style = "background=color:blue;padding:13px">
  <h1 style = "color:black;text-align:center;">Diabetes Prediction</h1>
  </div>
  """

  st.markdown(html_temp, unsafe_allow_html=True)

  Pregnancies = st.number_input("Pregnancies_Class: ")
  Glucose = st.number_input('Glucose_Class:')
  Blood Pressure = st.number_input("BloodPressure_Class: ")
  Skin Thickness = st.number_input("SkinThickness: ")
  Insulin = st.number_input("Insulin_Class: ")
  BMI = st.number_input("BMI_Class: ")
  Diabetes Pedigree Function = st.number_input("DiabetesPedigreeFunction: ")
  Age = st.number_input("Age_Class: ")

  result = ""

  if st.button("Show Prediction"):
    result = prediction_(Pregnancies_Class, Glucose_Class, BloodPressure_Class, SkinThickness, Insulin_Class, BMI_Class, DiabetesPedigreeFunction, Age_Class)
    st.success("Your Diabetes Prediction is {}".format(round(result[1])))

if __name__ == "__main__":
  main()
```
